```python
# Scrapping 7 different types of Cryptocurrency
BTC = api.search('bitcoin',count=200, lang='en', tweet_mode='extended')
ETH = api.search('Ethereum', count=200, lang='en', tweet_mode='extended')
USDT = api.search('tether', count=200, lang='en', tweet_mode='extended')
XRP = api.search('Ripple', count=200, lang='en', tweet_mode='extended')
BCH = api.search('bitcoin cash', count=200, lang='en', tweet_mode='extended')
LTC= api.search('litecoin', count=200, lang='en', tweet_mode='extended')
VTC = api.search('vertcoin', count=200, lang='en', tweet_mode='extended')
```
